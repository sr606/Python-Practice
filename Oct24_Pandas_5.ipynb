```python
import pandas as pd
import numpy as np
data = pd.read_csv('Pfizer_1.csv')
data.head()
```

```python
data_melt = pd.melt(data,id_vars = ['Date', 'Drug_Name', 'Parameter'],
                    var_name = "time",
                    value_name = "reading")
```

```python
data_melt.head()
```

```python
data_tidy = data_melt.pivot(index = ['Date','time','Drug_Name'],
                            columns = 'Parameter',
                            values = 'reading')

data_tidy.head()
```
