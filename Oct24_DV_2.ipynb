```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv('vgsales.csv')
data.head()
```

```python
data.describe()
```

##*Continuous Continuous*


```python
# How can we plot the sales trend over the years for the longest running game?
#first need to find out longest running game
game_life = data.groupby('Name').agg(min_year = ('Year','min'), max_year = ('Year','max'))
game_life['range'] = game_life['max_year'] - game_life['min_year']
game_life.sort_values(['range'], ascending= False)[:5]
```

```python
# lets query longest running game ice hockey
ih = data.loc[data['Name'] == 'Ice Hockey']
ih
```

```python
plt.title('Ice Hockey Sales Trend')
sns.lineplot(data = ih, x = 'Year', y = 'NA_Sales')
plt.ylabel('Sales')
plt.show()
```

```python
plt.title('Ice Hockey Sales Trend')
sns.lineplot(data = ih, x = 'Year', y = 'NA_Sales', color = 'r')
plt.ylabel('Sales')
plt.show()
```

```python
plt.title('Ice Hockey Sales Trend')
sns.lineplot(data = ih, x = 'Year', y = 'NA_Sales')
plt.ylabel('Sales')
plt.xlim(left = 1990, right = 2010)
plt.show()
```

```python
baseball = data.loc[data['Name'] == 'Baseball']
baseball.head()
```

```python
sns.lineplot(data = ih, x = 'Year', y = 'NA_Sales')
sns.lineplot(data = baseball, x = 'Year', y = 'NA_Sales')
plt.show()
```

```python
sns.lineplot(data = ih, x = 'Year', y = 'NA_Sales', label = 'Ice Hockey')
sns.lineplot(data = baseball, x = 'Year', y = 'NA_Sales', label = 'Baseball')
plt.show()
```

```python
# upper center
# upper left
# upper right
# lower right

sns.lineplot(data = ih, x = 'Year', y = 'NA_Sales', label = 'Ice Hockey')
sns.lineplot(data = baseball, x = 'Year', y = 'NA_Sales', label = 'Baseball')
plt.legend(loc = 'lower right')
plt.show()
```

```python
sns.lineplot(data = ih, x = 'Year', y = 'NA_Sales', label = 'Ice Hockey')
sns.lineplot(data = baseball, x = 'Year', y = 'NA_Sales', label = 'Baseball')
plt.legend(loc = 'lower right',ncols = 2)
plt.show()
```

```python
sns.lineplot(data = ih, x = 'Year', y = 'NA_Sales', label = 'Ice Hockey')
sns.lineplot(data = baseball, x = 'Year', y = 'NA_Sales', label = 'Baseball')
plt.legend(loc = 'lower right',ncols = 2, frameon = False)
plt.grid()
plt.show()
```

```python
top3_pub = data['Publisher'].value_counts().index[:3]
top3_gen = data['Genre'].value_counts().index[:3]
top3_plat = data['Platform'].value_counts().index[:3]

top3_data = data.loc[(data['Publisher'].isin(top3_pub)) & (data['Platform'].isin(top3_plat)) & (data['Genre'].isin(top3_gen))]
top3_data
```

```python
data["Platform"].value_counts()
```

```python
sns.countplot(x = 'Publisher', hue = 'Publisher', data = top3_data)
```

```python
sns.countplot(x = 'Publisher', hue = 'Platform', data = top3_data)
```
