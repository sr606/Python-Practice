central limit theorem relies on the concept of a **sampling distribution**, which is the probability distribution of statistic for a large number of samples taken frm population






**sampling distribution** = draw random samples frm population, calculate means for each sample, and repeat.

The collection of these means forms a sampling distribution.


image.png

The **Central Limit theorem** states that "the mean of random sample will resemble even closer to the population mean as sample size increases and it will appropriate normal distribution regardless of shape of population distribution"


Means if u take sufficiently large samples from a population, the samples means will be normally distributed, even if population isn't normally distributed.


image.png

image.png

we started by n= 2 from population, calculated their mean and repeated this many times to form a distribution.

**This distribution tendss to look more like a normal distribution.**


When we **incresed sample size, distribution resembled a normal distribution even more closely.**



A **smaller n (n=2), will high standard deviation** becoz sample means are less estimates of the population mean,
resulting in more spread.


A **large n (n> = 30) will have a low std dev** since sample means become more precise estimates of the population mean,

leading to less spread

image.png

#**Conclusion**




The **sample size not only influence how closely sampling distribution approximates a normal curve but also impacts the spread or precision of sample means.**


As the sample size increases,




*   CLT becoms more applicable
*   Std Dev decreases
*   Threby, making estimates of   population parameters more reliable.

#**CI**

Confidence Interval is mean of your estimate plus and minus the variation in that extimate


This is the range of values you expect your estimate to fall within a certain level of confidence.

what is confidence?



if you construct a confidence interval with 95% confidence level, u r confident that 95 out of 100 times estimate will fall between upper and lower values specified by cinfidence interval
