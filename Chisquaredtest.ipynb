```python
observed = [73, 38, 18]

# Calculate the expected values based on the proportions from the original survey data
total_survey_size = sum(observed)
expected = [0.60 * total_survey_size, 0.28 * total_survey_size, 0.12 * total_survey_size]

alpha = 0.05

chi_stat, p_value = chisquare(observed, expected)

print("Chi-squared statistic:", chi_stat)
print("P-value:", p_value)

if p_value > alpha:
  print("Can't reject the null hypothesis. Your survey supports the original one.")
else:
  print("Reject the null hypothesis. Your survey does not support the original one.")
```

```python
import scipy.stats as stats

# Observed frequencies from your survey
observed = [73, 38, 18]

# Expected frequencies based on the original survey
total_surveyed = sum(observed)
expected = [0.60 * total_surveyed, 0.28 * total_surveyed, 0.12 * total_surveyed]

# Perform the chi-squared test
chi_squared, p_value = stats.chisquare(f_obs=observed, f_exp=expected)

# Set the significance level
alpha = 0.05

# Determine the critical chi-squared value (2 degrees of freedom)
critical_chi_squared = stats.chi2.ppf(1 - alpha, df=2)

# Print the results
print(f"Chi-Squared Statistic: {chi_squared}")
print(f"P-Value: {p_value}")
print(f"Critical Chi-Squared Value: {critical_chi_squared}")

# Make a decision
if chi_squared > critical_chi_squared:
  print("Reject the null hypothesis")
else:
  print("Fail to reject the null hypothesis")
```
