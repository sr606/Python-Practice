Q: Square of numbers of a list using list comprehension

```python
a = [3,6,1,4]

ans = []

n = len(a)
for i in range(n):
  ans.append(a[i]**2)
```

```python
ans
```

```python
ans = [a[i]**2 for i in range(len(a))]
```

```python
ans
```

```python
#square the elements present at even index and dont do anything for the elements present at odd index
```

```python
a = [3,6,1,4]
#op: [9,6,1,4]
```

```python
ans = [a[i]**2 if i%2 == 0 else a[i] for i in range(len(a))]
```

```python
ans
```
